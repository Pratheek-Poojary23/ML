```python
# sigmoid function
import math

def sigmoid(x):
    return 1/ (1+ math.exp(-x))


print(sigmoid(100))
print(sigmoid(1))
```

```python
# tanh Fuction

def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

print(tanh(100))
print(tanh(-56))
```

```python
# relu functaion

def relu(x):
    return max(0,x)

print(relu(10))
print(relu(-5))
```

```python
# leaky relu
def leakyrelu(x):
    return max(0.1*x, x)

print(leakyrelu(-10))
```
