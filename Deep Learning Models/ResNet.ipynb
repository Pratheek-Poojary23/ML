```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
```

```python
x_train[0]
```

```python
#  x_train = tf.image.resize(x_train, [224, 224])
#  x_test = tf.image.resize(x_test, [224, 224])
#means you're trying to resize the entire training dataset (50,000 images) at once,
# which causes an Out of Memory (OOM) error — especially on CPU or limited GPU environments.
```

```python
# Define resize and preprocess function
def preprocess(image, label):
    image = tf.image.resize(image, [224, 224])
    return image, label

# Create tf.data.Dataset pipelines
BATCH_SIZE = 32

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_ds = train_ds.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test))
test_ds = test_ds.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

Pre trained model

```python
#feature_extractor_url = "https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/5"

feature_extractor_url= "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
feature_extractor_layer = hub.KerasLayer(
    feature_extractor_url,
    input_shape=(224, 224, 3),
    trainable=False  # Freeze weights
)
```

```python
model = tf.keras.Sequential([
    feature_extractor_layer,
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')  # CIFAR-10 has 10 classes
])

model.compile(
    optimizer='SGD',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
model.fit(train_ds, validation_data=test_ds, epochs=3, batch_size=4)
```
