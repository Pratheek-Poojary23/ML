```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset = pd.read_csv(r'C:\Users\PRATHEEK POOJARY\OneDrive\Desktop\Machine Learning\datasets\Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.info()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)

dataset['Landsize'] = dataset['Landsize'].fillna(dataset['Landsize'].mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset= pd.get_dummies(dataset,drop_first=True,dtype=int,)
```

```python
for col in dataset:
    print(f"{col}: {dataset[col].values}")
```

```python
dataset.head()
```

```python
x=dataset.drop(['Price'],axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
model.score(x_train,y_train)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

Here training score is 68% but test score is 13.85% which is very low

Normal Regression is clearly overfitting the data, let's try other models
Using Lasso (L1 Regularized) Regression Model

```python
from sklearn import linear_model
modelL1 = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
modelL1.fit(x_train, y_train)
```

```python
y_predL1=modelL1.predict(x_test)
```

```python
modelL1.score(x_train,y_train)
```

```python
modelL1.score(x_test,y_test)
```

```python
mse = mean_squared_error(y_test, y_predL1)
r2 = r2_score(y_test, y_predL1)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

Using Ridge (L2 Regularized) Regression Model

```python
from sklearn.linear_model import Ridge
modelL2= Ridge(alpha=50, max_iter=100, tol=0.01)
modelL2.fit(x_train, y_train)
```

```python
modelL2.score(x_train,y_train)
```

```python
modelL2.score(x_test,y_test)
```

```python
y_predL2=modelL2.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_predL2)
r2 = r2_score(y_test, y_predL2)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```
