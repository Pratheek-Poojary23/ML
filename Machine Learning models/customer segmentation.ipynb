```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv(r"C:\Users\PRATHEEK POOJARY\OneDrive\Desktop\Machine Learning\datasets\customer segment.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
gender_encoder= LabelEncoder()
marrage_encoder= LabelEncoder()
graduate_encoder=LabelEncoder()
spending_encode=LabelEncoder()
var_encoder= LabelEncoder()
profession_encoder= LabelEncoder()
segmentation_encoder= LabelEncoder()

df['Gender_e']=gender_encoder.fit_transform(df[['Gender']])
df['ever_married_e']=marrage_encoder.fit_transform(df[['Ever_Married']])
df['graduated_e']=graduate_encoder.fit_transform(df[['Graduated']])
df['spending_score_e']=spending_encode.fit_transform(df[['Spending_Score']])
df['var_1']=var_encoder.fit_transform(df[['Var_1']])
df['profession_e']=profession_encoder.fit_transform(df[['Profession']])
df['segmentation_e']=segmentation_encoder.fit_transform(df['Segmentation'])
```

```python
df=df.drop(columns=['Gender','Ever_Married','Graduated','Spending_Score','Var_1','Profession','Segmentation'],axis=1)
df.head()
```

```python
df.describe()
```

```python
x=df[['Age','profession_e']]
```

```python
from sklearn.cluster import KMeans

inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.plot(k_range, inertia, 'bo-')
plt.xticks(np.arange(0,11,1))
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.show()
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4)
y_pred=model.fit_predict(x)
df['cluster']=y_pred
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4=  df[df.cluster==3]

plt.scatter(df1['Age'],df1['profession_e'],color='blue')
plt.scatter(df2['Age'],df2['profession_e'],color='green')
plt.scatter(df3['Age'],df3['profession_e'],color='yellow')
plt.scatter(df4['Age'],df4['profession_e'],color='red')

plt.legend()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette='Set1')
plt.title('Customer Segmentation via K-Means Clustering')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(x, kmeans.labels_)
print(f"Silhouette Score: {score:.3f}")
```

```python
from sklearn.metrics import davies_bouldin_score

dbi = davies_bouldin_score(x, kmeans.labels_)
print(f"Davies-Bouldin Index: {dbi:.3f}")
```
