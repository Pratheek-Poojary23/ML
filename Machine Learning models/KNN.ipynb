```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\PRATHEEK POOJARY\OneDrive\Desktop\Machine Learning\datasets\Iris.csv")
data
```

```python
x=data.drop('Species',axis=1)
y=data['Species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.30)
```

```python
model= KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(8, 6))
plt.style.use("dark_background")
sns.scatterplot(data=data, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set1', s=70,)
plt.title('Iris Dataset - Sepal Length vs Sepal Width')
plt.legend(title='Species')
plt.show()
```

```python
cm= confusion_matrix(y_test,y_pred)
print(cm)
sns.heatmap(data=cm, annot=True, cmap='Blues')
```
