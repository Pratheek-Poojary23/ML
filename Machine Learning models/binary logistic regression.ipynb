```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\PRATHEEK POOJARY\OneDrive\Desktop\Machine Learning\datasets\logistic_data.csv")
df.head()
df.shape
```

```python
plt.yticks(np.arange(0,1.1,0.5))
plt.xlabel("Age")
plt.ylabel("Insurance bought")
plt.scatter(df.age, df.insurance)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df[['age']]
y=df['insurance']
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model= LogisticRegression()
model.fit(x_train,y_train)
```

```python
print(model.coef_)
model.intercept_                    #y_pred= 1/1+e^-(wx+b)
```

```python
x_test
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
plt.plot(x_test, y_test, 'ro')
plt.plot(x_test, y_pred, 'go')
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
cm= confusion_matrix(y_test, y_pred)
sns.heatmap(data=cm, annot=True,linewidths=15)
```
